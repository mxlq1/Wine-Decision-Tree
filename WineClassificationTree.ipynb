```python
from sklearn import tree, datasets
from sklearn.tree import plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np
import matplotlib.pyplot as plt
```

```python
d = datasets.load_wine()
```

```python
X_train, X_test, y_train, y_test = train_test_split(d['data'], d['target'], test_size=0.2)
```

```python
fig = plt.figure(figsize=(10, 8))

for depth in range(3, 7):
  model = tree.DecisionTreeClassifier(max_depth=depth, criterion='gini')
  model.fit(X_train, y_train)
  y_pred = model.predict(X_test)

  accuracy = accuracy_score(y_test, y_pred)

  #print(y_pred)
  #print(accuracy)

  ax = fig.add_subplot(2, 2, depth-2)
  ax.plot(y_test, label='Реальные значения', linestyle='dashed', c='b')
  ax.plot(y_pred, label='Предсказанные значения', c='r')
  ax.set_title(f'Accuracy = {accuracy}')

plt.tight_layout()
plt.show()
```

Увеличение глубины дерева не ведет к увеличению эффективности алгоритма классификации. При повторном запуске дерево может выдавать точность в 90% как при глубине 3, так и при глубине 6.
